Por baixo deste bloco existe um bloco de código (code cell), no momento em que na "toolbar" o botão indicar "Connected", seleciono bloco e executa-o.
Existem outras formas de executar um bloco de código:

- **Cmd/Ctrl+Enter** para executar o bloco corrente;
- **Shift+Enter**  executa o bloco corrente e salta para o proximo bloco, adicionando um novo se não existir; <-- *O mais usado *
- **Alt+Enter** executa o bloco e adiciona um novo bloco de código logo por baixo;

```
import requests
import numpy as np
import pandas as pd
import json
import datetime
from matplotlib import pyplot as plt

text = requests.get('https://min-api.cryptocompare.com/data/histohour?fsym=LUX&tsym=BTC&limit=60&aggregate=3&e=CCCAGG')

pd = pd.DataFrame(text.json()['Data'])
pd['timestamp'] = [datetime.datetime.fromtimestamp(d) for d in pd.time]

plt.title("LUX/BTC")
plt.plot(pd.timestamp, pd.close)
```
